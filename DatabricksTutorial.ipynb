```
# Transformation is when assigning a value to a variable which based on RDD it will create a copy of the original file where the original file #is imutable but the copy of the file is mutable.

# Action on the other hand is when we execute an operation on the record and it will take few time to execute the command

# Note: The above data are built in dataset

# Data in right menue help us to add data to the database
# Dataset can be add but should be enable first [Setting -> Admin Console -> Workspace Setting -> DBFS File Browser: (set to enable)]
    # Note: Referesh page to take effect

# Data if you now check data menue you will see two tab now 'Database Tables' and 'DBFS'

# To upload dataset in databricks within Notebook click [ file -> upload; upload dataset; copy the path from opened window

# Download dataset from www.kaggle.com
```

```
bike_sharing_data.select('season', 'holiday', 'cnt')\
    .limit(3)\
    .show()
```

```
bike_sharing_data.filter(bike_sharing_data['season'].isNull())\
    .show()
```

```
bike_sharing_data = bike_sharing_data.dropna()

bike_sharing_data.count()
```

```
bike_sharing_data_copy = bike_sharing_data

bike_sharing_data_copy.select('season', 'holiday', 'cnt')\
    .withColumnRenamed('season', 'season_ahmad')\
    .withColumnRenamed('holiday', 'holiday_ahmad')\
    .show(3)
```

```
bike_sharing_data_copy.select('season', 'holiday', 'cnt')\
    .withColumn("season_copy", bike_sharing_data_copy['season'])\
    .show(3)
```

```
bike_sharing_data_copy.select('season', 'holiday', 'cnt')\
    .orderBy(bike_sharing_data_copy['cnt'].asc())\
    .show(3)
```

```
bike_sharing_data_copy.groupBy('cnt').count().show(5)
```

```
bike_sharing_data_copy.groupBy('cnt').agg({'season': 'sum'}).show(5)
```

```
bike_sharing_data_copy.groupBy('cnt').count().write.csv('dbfs:/FileStore/shared_uploads/ahmad/ahmad_test.csv')
```
